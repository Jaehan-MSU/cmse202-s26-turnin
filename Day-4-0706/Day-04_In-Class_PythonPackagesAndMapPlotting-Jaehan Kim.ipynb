<font size=+3>&#9998;</font> This code makes a choropleth map of U.S. unemployment rates by state.

&#9989; **Question 6:** What controls in binning of values on the "Unemployment Rate" bar (upper right of the figure)? Can you set the bins manually? How would you do this? You might want to test it out with a bit of code or consult the folium documentation.

````python
import folium
import pandas as pd

state_geo = "us-states.json"
state_unemployment = "US_Unemployment_Oct2012.csv"

state_data = pd.read_csv(state_unemployment)

bins_map = folium.Map(location=[48, -102], zoom_start=3)

folium.Choropleth(
    geo_data=state_geo,
    data=state_data,
    columns=["State", "Unemployment"],
    key_on="feature.id",
    fill_color="YlGnBu",
    bins=[0, 4, 6, 8, 10, 12],
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="Unemployment Rate (%)",
    highlight=True
).add_to(bins_map)

bins_map
````

<font size=+3>&#9998;</font> The bins argument controls the binning of values on the unemployment rate color bar. I can set the bins manually by giving a list of bin edges, such as bins=[0, 4, 6, 8, 10, 12].
